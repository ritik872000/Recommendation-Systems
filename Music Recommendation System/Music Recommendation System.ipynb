# Merging both datasets into one

# Mapping the song name with its respective artist and inserting it into a new column

```python
song_df['song'] = song_df['title'].map(str) + " - " + song_df['artist_name']
```

```python
song_df.head(5)
```

# Grouping the songs based on there ascending order of listencount

```python
combined_song_listencount=song_df.dropna(axis=0,subset=['song'])
song_grouped=(combined_song_listencount.groupby(by=['song'])['listen_count'].
             count().
              reset_index().
             rename(columns={'listen_count':'totallistencount'})
              [['song','totallistencount']]
             )

song_grouped.head()
```

# Merging the dataset with totallistencount with old data

```python
rating_popular_song = combined_song_listencount.merge(song_grouped, left_on = 'song',right_on='song', how = 'left')
rating_popular_song.head()
```

# Setting the popularity threshold

```python
import seaborn as sns
sns.boxplot(y=rating_popular_song["totallistencount"])
```

# As we can see our centralized data in above plot I've calculated the median of our centralized data using tableau which is 434 so I've set the popularity threshold as 434 because median doesn't gets afected by outliers

```python
popularity_threshold=434
```

# Selecting songs which have listencount greater tham or equal to 434

```python
rating_popular_song= rating_popular_song.query('totallistencount >= @popularity_threshold')
rating_popular_song.head()
```

```python
rating_popular_song.shape
```

# Creating a pivot table to see the listen count of songs directly

```python
song_features_df=rating_popular_song.pivot_table(index='song',columns='user_id',values='listen_count').fillna(0)
```

```python
song_features_df.head(20)
```

# Modelling and fitting the K-NN model

```python
from scipy.sparse import csr_matrix

song_features_df_matrix = csr_matrix(song_features_df.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric='cosine', algorithm='brute')
model_knn.fit(song_features_df_matrix)
```

```python
print(song_features_df_matrix)
```

# Making a Recommender

```python
from fuzzywuzzy import process
```

```python

def recommender(song_name, data,model, n_recommendations ):
    model.fit(data)
    idx=process.extractOne(song_name, song_df['song'])[2]
    print('Song Selected: ',song_df['song'][idx], 'Index: ',idx)
    print('Searching for recommendations.....')
    distances, indices=model.kneighbors(data[idx], n_neighbors=n_recommendations)
    for i in indices:
        print(song_df['song'][i].where(i!=idx))
    
```

# Finally! the recommendations for given song

```python
recommender('yellow-coldplay',song_features_df_matrix,model_knn,15)
```
